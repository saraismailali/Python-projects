```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
numeric= T_clean[['Title','Years','FamilySize','Southampton(S)','Cherbourg(C)','Queenstown(Q)','Pclass','Survived']]
x_train ,  x_test , y_train  , y_test = train_test_split(numeric.drop('Survived',axis=1),numeric['Survived'])
LogReg = LogisticRegression(solver='saga')
LogReg.fit(x_train,y_train)
```

```python
accuracy_score(y_train,LogReg.predict(x_train))
```
